```python
# Economic statistics - Regions
agencyID = 'OECD.CFE.EDS'
dataflowID = 'DSD_REG_ECO@DF_ECO'
```

```python
# create a Structure instance
s = notoecd.get_structure(agencyID, dataflowID)
```

```python
s.toc # table of contents
```

```python
s.explain_vals('MEASURE') # dict of code:label
```

```python
filters = {
    'territorial_level': ['tl2', 'tl3'], # lists/tuples/sets for multiple values
    'measure': 'Gross domestic product', # labels or codes, both work
    'PRICES': 'Q', 
    'unit_measure': 'USD_ppp_PS' # case insensitive
}
```

```python
df = notoecd.get_df(
    agencyID = agencyID, 
    dataflowID = dataflowID, 
    filters = filters,
    startYear = 2020,
    endYear = 2022
)
```

```python
df.head()
```

```python
# use Structure to explain columns
column_dict = s.explain_columns(df)
print(column_dict['REF_YEAR_PRICE'])
```
